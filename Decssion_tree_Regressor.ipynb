```python
import matplotlib.pyplot as plt 
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeRegressor
```

```python
df = load_diabetes(as_frame=True).frame
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X_train,X_test, y_train, y_test = train_test_split(
    X,y, random_state=42, test_size=0.2
)
```

```python
model = DecisionTreeRegressor(max_depth=9, min_samples_leaf=30)
model.fit(X_train, y_train)
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)
```

```python
print("MAE_traning: ", mean_absolute_error(y_train, y_pred_train))
print("MAE_test: ", mean_absolute_error(y_test, y_pred_test))

print("r^2_traning: ", r2_score(y_train, y_pred_train))
print("r^2_test: ", r2_score(y_test, y_pred_test))
```

```python
plt.figure(figsize=(15,8))
plot_tree(
    model,
    feature_names=X.columns,
    filled=True,
    # max_depth=2
)
plt.tight_layout()
plt.show()
```
